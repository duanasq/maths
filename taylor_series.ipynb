```python
def taylor_series_for_e(a, x, n):
    sum = 0
    for i in range(n):
        sum += e^a * x^i / factorial(i)
    return sum

    
```

```python
for i in range(1,21):
    res = taylor_series_for_e(0, 1, i)
    print(f'n={i:<5}{str(res):<35}{N(res)}')

e_digits = N(e^1)
print(f'{"e is ":<42}{e_digits}')
```
